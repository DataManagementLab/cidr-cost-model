# Data Efficiency

## For Scan


```python
import matplotlib.pyplot as plt
import pandas as pd

fig = plt.figure(figsize=(10,5))

df_black_box = pd.read_csv("plots/black_box/black_box_data_efficiency.csv")
df_black_box['training_data_size'] = df_black_box['train_size']
plt.plot(df_black_box['training_data_size'], df_black_box['q_error_median'], marker='o', label='Black Box')

df_white_box = pd.read_csv("plots/white_box/white_box_data_efficiency.csv")
# divide by 10 because no inflation is required
df_white_box['training_data_size'] = df_white_box['train_size'] / 10

# adding more data would not change anything, hence extrapolate last median q-error
training_data_sizes = list(df_white_box['training_data_size']) + [100]
last_q_error = df_white_box[df_white_box.train_size == 80.0]['q_error_median'].sum()
q_errors = list(df_white_box['q_error_median']) + [last_q_error]

plt.plot(training_data_sizes, q_errors, marker='o', label='White Box')

plt.legend()
plt.ylabel('Median q-error')
plt.xlabel('Training Data Size ()') 
```

## For Join


```python
import matplotlib.pyplot as plt
import pandas as pd

fig = plt.figure(figsize=(10,5))

df_black_box = pd.read_csv("plots/black_box/query_plan_stats/query_plan_stats_data_efficiency.csv")
df_black_box['training_data_size'] = df_black_box['train_size']
plt.plot(df_black_box['training_data_size'], df_black_box['q_error_median'], marker='o', label='Black Box')

df_white_box = pd.read_csv("plots/white_box/white_box_join_data_efficiency.csv")
# divide by 10 because no inflation is required
df_white_box['training_data_size'] = df_white_box['train_size'] / 10

# adding more data would not change anything, hence extrapolate last median q-error
training_data_sizes = list(df_white_box['training_data_size']) + [100]
last_q_error = df_white_box[df_white_box.train_size == 10.0]['q_error_median'].sum()
q_errors = list(df_white_box['q_error_median']) + [last_q_error]

plt.plot(training_data_sizes, q_errors, marker='o', label='White Box')

plt.legend()
plt.ylabel('Median q-error')
plt.xlabel('Training Data Size (%)') 
```
